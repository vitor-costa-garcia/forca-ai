```python
from src.Agent import Agent
from src.Environment import Environment
from src.State import State

NUM_EPISODES = 500
```

Inicializando o ambiente

```python
agent = Agent()
env = Environment(5)

agent.load('src/data/words.txt')
env.load('src/data/words.txt')
```

```python
env.reset()
agent.reset(env.getState().getWordLen())
```

```python
winCounter = 0
winRate = 0
winCounterLastFH = 0

for episode in range(NUM_EPISODES):
    gameState = 0
    if (episode+1)%500 == 0:
        print(f"Episode: {episode+1}")
        print(f"Current total winrate: {winRate} | Last 500 episodes winrate: {winCounterLastFH/500} | Current wincounter: {winCounter}")
        winCounterLastFH = 0
    while gameState == 0:

        currentState = env.getState()
        agentGuess = agent.guess(currentState)
        env.step(agentGuess)
        gameState = env.getGameSituation()

    realWord = env.getRealWord()
    if gameState == 1:
        winCounter += 1
        winCounterLastFH += 1
    else:
        print(f"Lost: {realWord}")

    winRate = winCounter/(1+episode)

    agent.learn(realWord)

    env.reset()
    agent.reset(env.getState().getWordLen())

```
